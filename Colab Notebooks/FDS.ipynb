```python
x = np.array([5,15,25,35,45,55]).reshape((-1,1))
y = np.array([5,20,14,32,22,38])
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
r_sq = model.score(x,y)
print(f"coefficient of determination:{r_sq}")
```
